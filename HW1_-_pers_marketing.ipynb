```python
#подгружаем movies
movies=pd.read_csv('movies.csv', sep=',')
movies
```

```python
#подгружаем ratings
ratings=pd.read_csv('ratings.csv', sep=',')
ratings
```

```python
#датасеты
movies = movies.merge(ratings, on = 'movieId', how = 'inner')
movies
```

```python
#оставляем только нужные нам переменные
movies_SVD = movies[['userId','movieId','rating']]
movies_SVD
```

```python
from surprise import Dataset, Reader
reader=Reader(rating_scale=(1,5))
data = Dataset.load_from_df(movies_SVD, reader)
```

```python
from surprise import SVD #подгружаем методом SVD матрицы
algo = SVD(n_factors=100, n_epochs=100, lr_all=0.005, reg_all=0.1)
```

```python
from surprise.model_selection import cross_validate
cross_validate(algo, data, measures=['MSE', 'MAE'], cv=5, verbose=True)
```

Наше MSE=0.7224
MAE=0.6510
